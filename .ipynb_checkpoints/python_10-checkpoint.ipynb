```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b, c, d, e = -12, -18, 5, 10, -30
```

```python
def f(x):
    function = a*x**4 * np.sin(np.cos(x)) + b*x**3 + c*x**2 + d*x + e
    return function
```

```python
x = np.arange(-100, 100.01, 0.01)
```

```python
def show_func():
    plt.plot(x, f(x), 'r', label = "функкция f(x)")
    plt.show()
```

```python
print(show_func())
```
